# Bike Sharing Dataset: Column Explanations

This dataset contains hourly records of bike rentals along with corresponding weather and seasonal information. Below is a detailed breakdown of what each column represents:

| Column Name | Description | Data Type |
| :--- | :--- | :--- |
| **datetime** | The specific date and hour the data was recorded (e.g., `2011-01-01 00:00:00`). | Datetime |
| **season** | The current season of the year at the time of recording (e.g., `Spring`, `Winter`). | String |
| **holiday** | A binary indicator showing whether the day is a recognized public holiday (`1.0` = Yes, `0.0` = No). | Float (Boolean) |
| **workingday** | A binary indicator showing whether the day is a standard working day (`1.0` = Yes, `0.0` = No). A value of `0.0` usually indicates a weekend or holiday. | Float (Boolean) |
| **weather** | A categorical description of the weather condition during that hour (e.g., `Clear`, `Mist`). | String |
| **temp** | The actual temperature recorded during that hour (typically in Celsius). | Float |
| **humidity** | The relative humidity level as a percentage. | Float |
| **windspeed** | The recorded wind speed during that hour. | Float |
| **casual** | The count of bike rentals made by unregistered/casual users during that hour. | Integer |
| **registered** | The count of bike rentals made by registered/subscribed users during that hour. | Integer |
| **rented_bikes_count** | The total number of bikes rented during that hour. This is the sum of the `casual` and `registered` columns. | Integer |

---
*Note: Some rows contain missing values (represented as blanks/NaNs) which may require imputation or cleaning during exploratory data analysis.*

```python
df.dropna(subset='season',inplace=True,ignore_index=True)
```

```python
df
```

```python
#Fill missing values in columns (holiday, workingday, weather) with the mode

df['holiday']=df['holiday'].fillna(df['holiday'].mode)
df['workingday']=df['workingday'].fillna(df['workingday'].mode)
df['weather']=df['weather'].fillna(df['weather'].mode)



```

```python
df.isna().sum()
```

```python
#Fill missing values in column (humidity) with the median
df['humidity']=df['humidity'].fillna(df['humidity'].median())
```

```python
df.isna().sum()
```

```python
num_cols = df.select_dtypes(include= 'number').columns
```

```python
#Fill missing values in columns (temp, windspeed) with the KNN imputer
from sklearn.impute import KNNImputer
knn = KNNImputer()
df[num_cols]=knn.fit_transform(df[num_cols])


```
