### Bonus Question (Search for cumsum function)
### Q20. Cumulative revenue trend across the dataset timeline.
👉 Recommended visualization: **Line chart with cumulative sum**

```python
rev = df.groupby('Date')['money'].sum().reset_index()
```

```python
rev['cumulative_revenue'] = rev['money'].cumsum()
```

```python
# Write your visualization code here
px.line(rev,x='Date',y='cumulative_revenue',title='Cumulative Revenue Over Time')
```
